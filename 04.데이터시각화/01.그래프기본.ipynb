```python
import matplotlib.pyplot as plt
```

```python
x = [-1, 2, 3]
y = [2, 4, 8]
plt.plot(x, y)
plt.show()
```

```python
#한글설정
plt.rc('font', family = 'Malgun Gothic')
plt.rc('font', size = 20)
plt.rc('axes', unicode_minus = False )
```

```python
import matplotlib.font_manager as fm
flist = fm.fontManager.ttflist
[f.name for f in flist]
```

```python
plt.plot(x, y)
plt.title('꺾은선 그래프')
plt.show()
```

```python
plt.rc('font', size = 10)
plt.title('꺾은선그래프', size = 20)
plt.plot(x, y)

plt.xlabel('X축', color = 'red', size = 10, loc = 'right')
plt.ylabel('Y축', color = 'blue', size = 10, loc = 'top')
plt.show()
```

```python
plt.plot(x, y)
plt.xticks([-2, 1, 3, 5])
plt.xticks(x)
plt.yticks(y)
plt.yticks([y for y in range(0, 11, 2)])
plt.show()
```
