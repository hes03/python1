```python
import pandas as pd
df = pd.read_csv('data/score.csv', index_col='지원번호')

import matplotlib.pyplot as plt

#한글설정
plt.rc('font', family = 'Malgun Gothic')
plt.rc('font', size = 10)
plt.rc('axes', unicode_minus = False )
```

```python
name = df['이름']
height = df['키']
```

```python
plt.barh(name, height)
plt.ylabel('학생이름')
plt.xlabel('키')
plt.xlim(160, 210)
plt.show()
```

```python
plt.figure(figsize=(10, 3))
bar = plt.bar(name, height, color = 'none', ec = 'g', hatch = '..')
plt.ylim(160, 210)
bar[0].set_hatch('/')
for idx, h in enumerate(height):
    plt.text(idx, h+2, f'{h}cm', ha='center')
plt.show()
```

```python
kor = df['국어']
plt.figure(figsize=(10, 3))
plt.bar(name, kor, width = 0.5)
plt.show()
```

```python
#학생별 평균을 막대그래프로 출력
#1) ylim(0-100)
#2) yticks(0, 5, 10, 15,,,,) 0부터 100까지
#3) xlabel 학생이름
#4) ylabel 평균점수
#5 학생별 막대위에 평균점수 출력
#6 그래프 제목 학생별 평균
#7. 그래프 사이즈 10, 5

df['평균'] = df.apply(lambda row:row['국어':'사회'].sum()/5, axis = 1)
df
avg = df['평균']

plt.figure(figsize=(10, 5))
plt.title('학생별 평균')
plt.bar(name, avg)
plt.xlabel('학생이름')
plt.ylabel('평균점수')
plt.ylim(0, 100)
plt.yticks([y for y in range(0, 101, 5)])
for idx, a in enumerate(avg):
    plt.text(idx, a+2, a, ha='center')

plt.show()
```

```python
import numpy as np
def graph(name):
    filt = df['이름'] == name
    df2 = df[filt]
    if len(df2) == 0:
        return np.nan
    else:    
        df3 = df2.loc[:,'국어':'사회']
        df4 = df3.iloc[0]
        plt.bar(df4.index, df4.values)
        return plt
    
    
image = graph('변덕규')
if pd.isna(image):
    print('없음')
else:
    plt.show()
```
