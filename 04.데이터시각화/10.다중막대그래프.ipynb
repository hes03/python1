```python
import matplotlib.pyplot as plt
plt.rc('font', family='Malgun Gothic')
plt.rc('axes', unicode_minus=False)

import pandas as pd
df = pd.read_csv('data/score.csv')
df
```

```python
x = df.index
x = x * 3
x
```

```python
x = df.index * 4
plt.bar(x+0, df['국어'], label='국어', width=1)
plt.bar(x+1, df['영어'], label='영어', width=1)
plt.bar(x+2, df['수학'], label='수학', width=1)
xticks = [idx+1 for idx in x] #bar 1개 늘어날때마다 0.5씩 중심 이동
plt.xticks(xticks, df['이름'])
for idx in df.index:
    x = idx * 4
    plt.text(x, df.loc[idx, '국어']/2, df.loc[idx, '국어'], ha='center', size=8)
    plt.text(x+1, df.loc[idx, '영어']/2, df.loc[idx, '영어'], ha='center', size=8)
    plt.text(x+2, df.loc[idx, '수학']/2, df.loc[idx, '수학'], ha='center', size=8)
plt.legend()
plt.show()
```

```python
x = df.index*3
plt.figure(figsize=(10, 4))
plt.bar(x, df['사회'], label='사회', width=1)
plt.bar(x+1, df['과학'], label='과학', width=1)
plt.legend(ncol=2)
plt.ylim(0, 100)
xticks = [idx+0.5 for idx in x]
plt.xticks(xticks, df['이름'])
for idx in df.index:
    x = idx * 3
    plt.text(x, df.loc[idx, '사회'], df.loc[idx, '사회'], ha='center', size=8)
    plt.text(x+1, df.loc[idx, '과학'], df.loc[idx, '과학'], ha='center', size=8)
plt.show()
```

```python
x = df.index * 4
plt.figure(figsize=(10, 5))
plt.bar(x, df['국어'], width=0.9, label='국어', color='lightgreen')
plt.bar(x+1, df['영어'], width=0.9, label='영어', color='skyblue')
plt.bar(x+2, df['수학'], width=0.9, label='수학', color='pink')
plt.legend()
xticks = [xtick+1 for xtick in x]
plt.xticks(xticks, df['이름'])
plt.show()
```
