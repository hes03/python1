```python
with open('영어가사.txt', 'r') as file:
    text = file.read()
print(text)
```

```python
from wordcloud import WordCloud
import matplotlib.pyplot as plt
```

```python
wc = WordCloud().generate(text)
```

```python
plt.imshow(wc, interpolation='bilinear')
plt.show()
```

```python
with open('paperplane.txt', 'r', encoding='utf-8') as file:
    text = file.read()
#print(text)
wc = WordCloud().generate(text)
plt.imshow(wc)
plt.axis('off')
plt.savefig('golden.png')
plt.show()
```

```python
import numpy as np
from PIL import Image
```

```python
heart_mask = np.array(Image.open('heart_mask.png'))
heart_mask
```

```python
wc = WordCloud(width=800, height=800, background_color='white', 
                min_font_size=20, mask=heart_mask).generate(text)
```

```python
plt.figure(figsize=(5, 5))
plt.imshow(wc, interpolation='bilinear')
plt.axis('off')
plt.savefig('paperplane.png')
plt.show()
```

```python
import konlpy
from konlpy.tag import Okt
from collections import Counter
```

```python
okt = Okt()
```

```python
with open('바이썸머.txt', 'r', encoding='utf-8') as file:
    text = file.read()
print(text)
```

```python
okt = Okt()
nouns = okt.nouns(text)
nouns = [noun for noun in nouns if len(noun)>1]
nouns
```

```python
counter = Counter(nouns)
counter
```

```python
counter = Counter(nouns)
```

```python
wc = WordCloud(width=800, height=800,
               font_path='MALGUN.TTF').generate_from_frequencies(counter)
```

```python
plt.imshow(wc)
plt.axis('off')
plt.show()
```
