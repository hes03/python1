```python
import pandas as pd
```

```python
data = pd.read_csv('공공보건의료기관현황.csv', encoding='CP949')
data
```

```python
data['시도'].unique()
```

```python
addr1 = data.loc[0, '주소']
addr1.split(' ')[:2]
```

```python
addr = data['주소'].apply(lambda row:row.split(' ')[:2]).tolist()
addr
```

```python
addr = pd.DataFrame(addr, columns=['시도', '군구'])
addr
```

```python
addr['시도'].unique()
```

```python
new_address = {'강원도':'강원특별자치도', '전라북도':'전북특별자치도'}
addr['시도'] = addr['시도'].apply(lambda x:new_address.get(x, x))
addr['시도'].unique()
```

```python
addr['시도군구'] = addr['시도'] + ' ' + addr['군구']
addr.head(2)
```

```python
addr_group = addr.groupby(['시도', '군구', '시도군구'], as_index=False).size()
addr_group.rename(columns={'size':'count'}, inplace=True)
addr_group.set_index('시도군구', inplace=True)
addr_group.head()
```

```python
population = pd.read_csv('행정구역별 인구수.csv', index_col=0)
population.rename(columns={'행정구역1':'시도', '행정구역2':'군구'}, inplace=True)
population['시도군구'] = population['시도'] + ' ' + population['군구']
population
```

```python
filt = population['군구'] != '합계'
population = population[filt]
filt = population['군구'] != '소계'
population = population[filt]
population.set_index('시도군구', inplace=True)
population
```

```python
addr_group.head(2)
```

```python
local_population = addr_group.merge(population, on='시도군구')
local_population
```

```python
cols = local_population.columns
local_population = local_population[['시도_x', '군구_x', 'count', '총인구수']]
local_population.rename(columns={'시도_x':'시도', '군구_x':'군구'}, inplace=True)
local_population['ratio'] = local_population['count'] / local_population['총인구수']*100000
local_population.head()
```

```python
local_population.to_csv('인구수별 공공의료기관수.csv')
```

```python
local = pd.read_csv('인구수별 공공의료기관수.csv', index_col=0)
local.sort_values('count', ascending=False)
```

```python
group = local.groupby('시도')['count'].sum()
group
```

```python
import matplotlib.pyplot as plt

#한글설정
plt.rc('font', family = 'Malgun Gothic')
plt.rc('font', size = 10)
plt.rc('axes', unicode_minus = False )
```

```python
plt.barh(group.index, group.values)
plt.show()
```

```python
filt = local['시도'] == '경기도'
local1 = local[filt]
local1.head()
```

```python
filt = local['시도'] == '인천광역시'
local1 = local[filt]
local1.head()
```

```python
def custom_autopct(pct):
    if pct > 3:
        return f'{pct:.0f}%'
    else:
        ''
```

```python
values = local1['count']
labels = local1['군구']
plt.pie(values, labels=labels, explode=[0.08]*len(local1), rotatelabels=True,
        autopct=custom_autopct, wedgeprops={'width':0.9, 'alpha':0.5, 'edgecolor':'white', 'linewidth':0.8}, textprops={'fontsize':9})
plt.show()
```
